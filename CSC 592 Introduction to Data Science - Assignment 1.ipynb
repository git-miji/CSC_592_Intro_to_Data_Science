```python
# Importing NumPy libraries for generating data class:

import numpy as np
from numpy import random

#Class1
mean1 = (random.uniform(0, 3), random.uniform(0,3), random.uniform(0,3))
cov1 = (np.diag((random.uniform(1, 10), random.uniform(0,10), random.uniform(0,10))))
points1 = np.random.randint(1, 100)
data1 = np.random.multivariate_normal(mean1, cov1, points1)

#Class 2
mean2 = (random.uniform(0, -4), random.uniform(0,-4), random.uniform(0,-4))
cov2 = (np.diag((random.uniform(1, 10), random.uniform(0,10), random.uniform(0,10))))
points2 = np.random.randint(1, 100)
data2 = np.random.multivariate_normal(mean2, cov2, points2)

# Class 3  
mean3 = (random.uniform(2, 4), random.uniform(2,4), random.uniform(2,4))
cov3 = (np.diag((random.uniform(1, 10), random.uniform(0,10), random.uniform(0,10))))
points3 = np.random.randint(1, 100)
data3 = np.random.multivariate_normal(mean3, cov3, points3)

# Class 4   
mean4 = (random.uniform(5, 7), random.uniform(5,7), random.uniform(5,7))
cov4 = (np.diag((random.uniform(1, 10), random.uniform(0,10), random.uniform(0,10))))
points4 = np.random.randint(1, 100)
data4 = np.random.multivariate_normal(mean4, cov4, points4)
        
# Putting all the data sets into a single vertical array:
X = np.vstack((data1, data1, data3, data4))

# Creating an array of labes for corresponding X Array:
y=np.hstack((np.ones(points1), np.ones(points2), np.ones(points3), np.ones(points4)))

# Importing Matplotlib, Pyplot and Axes3D for plotting the data classes:
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

#Create a 3-D Plot
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

# Defining color for each class:
ax.scatter(data1[:,0], data1[:,1],  data1[:,2], c='r')
ax.scatter(data2[:,0], data2[:,1],  data2[:,2], c='g')
ax.scatter(data3[:,0], data3[:,1],  data3[:,2], c='b')
ax.scatter(data4[:,0], data4[:,1],  data4[:,2], c='k')

#Setting limit in 3 Axes:
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_zlim(-10,10)

# Labelling Axes:
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_zlabel("Z Label")
plt.show()
```
