```python
#imports

from sklearn.datasets import make_moons
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

1) Use the make_moons dataset from scikit-learn with 500 samples, noise = 0.1 to make an artificial data-set. This is a 2 class data-set.

```python
data, labels = make_moons(n_samples=500, noise= 0.1)
print(data.shape)
print(labels.shape)
```

2) Split this into training and validation data.

```python
train_data = data[:400,:]
train_labels = labels[:400]
val_data = data[400:, :]
val_labels = labels[400:]
print(train_data.shape)
print(train_labels.shape)
```

3) Plot both the training and validation datasets.

```python
#platting training data
plt.rcParams["figure.figsize"] = (7,5)
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_labels)
plt.show()
```

```python
#plotting validation data
plt.rcParams["figure.figsize"] = (7,5)
plt.scatter(val_data[:, 0], val_data[:, 1], c=val_labels)
plt.show()
```

4) Implement a neural network model (that has reasonable performance) using the Sequential API.

```python
#building the model
model_sequential = keras.Sequential()
model_sequential.add(layers.Dense(16, activation="relu", name="first_layer"))
model_sequential.add(layers.Dense(8, activation="relu", name="second_layer"))
model_sequential.add(layers.Dense(4, activation="relu", name="third_layer"))
model_sequential.add(layers.Dense(2, activation="relu", name="fourth_layer"))
model_sequential.add(layers.Dense(1, activation="sigmoid", name="output_layer"))
model_sequential.build(input_shape=(None, 2))
```

```python
#compiling
model_sequential.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

#training the model using train data
history = model_sequential.fit(train_data,
                    train_labels,
                    epochs=200,
                    batch_size=128,
                    validation_data=(val_data, val_labels))
```

5) Plot the loss (of both training and validation) over the training epochs.

```python
#plotting training and validation loss loss values for each epoch
history_dict = history.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)
plt.rcParams["figure.figsize"] = (20,15)
plt.plot(epochs, loss_values, "x", label="Training loss")
plt.plot(epochs, val_loss_values, "b", label="Validation loss")
plt.title("Sequential API")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

6) Plot the data and the decision regions together on a new plot.

```python
res = 400

plt.rcParams["figure.figsize"] = (10, 8)

def addColor(values):
    colors = []
    for i in values:
        if i < 0.5:
            colors.append("blue")
        else:
            colors.append("orange")
    return colors

# Make meshgrid of xy points
xx, yy = np.meshgrid(np.linspace(-3, 3, res), np.linspace(-3, 3, res), indexing="xy")
xy = np.dstack((xx, yy))

# Make meshgrid of equal size containing predictions
zz = np.ndarray((res, res))
for i in range(0, res):
    zz[i] = model_sequential(xy[i])[:, 0]

# Graph decision regions
plt.contourf(xx, yy, zz, levels=8)

# predictions_sequential = np.round(model_sequential.predict(x_val))
xx, yy= np.meshgrid(data[:,0], data[:,1], sparse=True)
plt.scatter(xx, yy, c=addColor( model_sequential.predict(data) ) )
plt.title("Sequential API Decision Regions")
plt.show()
```

7) Repeat this using the functional API.

```python
#create the model
inputs = keras.Input(shape=(2,))
layer0 = layers.Dense(16, activation="relu")(inputs)
layer1 = layers.Dense(8, activation="relu")(layer0)
layer2 = layers.Dense(4, activation="relu")(layer1)
layer3 = layers.Dense(2, activation="relu")(layer2)
outputs = layers.Dense(1, activation="sigmoid")(layer3)
model_func = keras.Model(inputs=inputs, outputs=outputs)
```

```python
#compile
model_func.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

#train the model
history_func = model_func.fit(train_data,
                    train_labels,
                    epochs=400,
                    batch_size=128,
                    validation_data=(val_data, val_labels))
```

```python
#plotting training and validation loss loss values for each epoch
history_dict = history_func.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)
plt.rcParams["figure.figsize"] = (20,15)
plt.plot(epochs, loss_values, "x", label="Training loss")
plt.plot(epochs, val_loss_values, "b", label="Validation loss")
plt.title("Functional API")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
